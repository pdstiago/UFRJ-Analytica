**Análise do IDH Carioca**

```python
import pandas as pd
```

```python
adh=pd.read_csv('adh.csv')
adh.head()
```

```python
#filtro a tabela para o município do Rio e seleciono as colunas que irei analisar

adh_rio=adh[adh['id_municipio']==3304557]

colunas=['ano', 'idhm', 'idhm_e', 'idhm_l', 'idhm_r']

idh_rio=adh_rio[colunas]
idh_rio=idh_rio.dropna(axis=0)
```

```python
idh_rio
```

```python
#gráfico do IDH Educacional do Rio

idh_rio.plot(x='ano', y='idhm_e', title='IDH Educacional do Rio')
```
